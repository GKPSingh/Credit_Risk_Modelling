```python
# Importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
pd.set_option('display.float_format', lambda x: '{:.2f}'.format(x))
np.set_printoptions(suppress=True)
```

```python
# Loading the data
df_customers = pd.read_csv('customers.csv')
df_loans = pd.read_csv('loans.csv')
df_bureau = pd.read_csv('bureau_data.csv')
```

```python
# checking the size of the data
df_customers.shape, df_loans.shape, df_bureau.shape
```

```python
# Combining the different data files
df = pd.merge(df_customers, df_loans, on='cust_id')
df.head()
```

```python
# Checking the bureau data
df_bureau.head()
```

```python
# Merging the data
df = pd.merge(df, df_bureau, on='cust_id')
df.head()
```

```python
# Checking the shape of the merged data
df.shape
```

```python
# Getting the information about default column
df['default'] = df['default'].astype(int)
df.default.value_counts()
```

```python
# Getting the information 
df.head()
```

```python
# Splitting the data into train and test
X = df.drop('default', axis=1)
y = df['default']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.25, random_state=42)
```

```python
df_train = pd.concat([X_train, y_train], axis = 1)
df_test = pd.concat([X_test, y_test], axis = 1)
```

```python
# finding the continuous and categorical columns
columns_continuous = [ 'age', 'income', 'number_of_dependants', 'years_at_current_address', 
                       'sanction_amount', 'loan_amount','processing_fee', 'gst', 'net_disbursement', 
                      'loan_tenure_months','principal_outstanding', 'bank_balance_at_application',
                    'number_of_open_accounts','number_of_closed_accounts', 'total_loan_months', 'delinquent_months',
                   'total_dpd', 'enquiry_count', 'credit_utilization_ratio']
columns_categorical = [ 'gender', 'marital_status', 'employment_status','residence_type', 'city', 
                       'state', 'zipcode', 'loan_purpose', 'loan_type', 'default']
```

```python
# Plotting the box plot of continous columns 
import math

df_plot = df_train[columns_continuous]

n_cols = 3
n_rows = math.ceil(len(columns_continuous) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(22, 5 * n_rows))
axes = axes.flatten()

for i, col in enumerate(columns_continuous):
    axes[i].boxplot(
        df_plot[col].dropna(),
        vert=False,  # ✅ horizontal box plot
        patch_artist=True,
        boxprops=dict(facecolor='steelblue', alpha=0.7),
        medianprops=dict(color='navy', linewidth=2),
        whiskerprops=dict(color='steelblue'),
        capprops=dict(color='steelblue'),
        flierprops=dict(marker='o', alpha=0.4)
    )

    axes[i].set_title(col, fontsize=11, fontweight='bold')
    axes[i].grid(axis='x', linestyle='--', alpha=0.3)

# Hide unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

* Age is mostly between 30–50, with a few older outliers.
* Dependants are usually 0–3, median around 2.
* Years at address is typically 8–25 years, median ~15.
* Income is highly right-skewed with a few extremely high earners.
* Loan amount, sanction amount, and net disbursement are also very skewed.
* Most borrowers have small–medium loans, few have very large ones.
* Processing fee and GST are mostly small with rare big values.
* Bank balance is low for most, very high for a few.
* Principal outstanding is mostly moderate, with some extreme cases.
* Loan tenure is usually 15–35 months, median ~25.
* Total loan months centers around 40–100, with some very long histories.
* Open accounts are mostly 1–3; closed accounts 0–2.
* Enquiry count is typically 3–6.
* Delinquency and total DPD are low for most but high for some risky cases.
* Credit utilization is mostly 20–60%, with some near 100%.
* Overall, financial variables are heavily skewed with many outliers.
* Data needs outlier handling or transformation before modeling.

## Exploratory Data Analysis

```python
# Plotting kde plot for age with hue by default
sns.kdeplot(df_train_1[df_train_1.default==0]['age'], fill =True, label='Default=0')
sns.kdeplot(df_train_1[df_train_1.default==1]['age'], fill =True, label='Default=1')
plt.title(f'Age KDE Plot with Hue by default')
plt.legend()
plt.show()
```

```python
# Plotting kde plot for all continuous columns with hue by default
plt.figure(figsize=(24,20))    # width, height in inches

for i, col in enumerate(columns_continuous):
    plt.subplot(6, 4, i+1)  # 1 row, 4 columns, ith subplot
    sns.kdeplot(df_train_1[col][df_train_1.default==0], fill =True, label='default=0')
    sns.kdeplot(df_train_1[col][df_train_1.default==1], fill =True, label='default=1')
    plt.title(col)
    plt.xlabel('')
plt.tight_layout()
plt.show()
```

* The figure shows KDE (density) curves for each continuous feature, split by default status.

* Blue represents non-defaulters (default=0) and orange represents defaulters (default=1).

* For age, both groups are centered around 35–45 with only minor differences.

* Number of dependants shows discrete peaks, with similar shapes for both groups.

* Years at current address is fairly flat and similar across groups.

* Income and all loan-size variables are heavily right-skewed for both groups.

* Most customers are in the low–mid range, with a long tail of large values.

* Sanction amount, loan amount, and net disbursement almost fully overlap.

* Processing fee and GST follow the same skewed pattern.

* Bank balance at application is low for most, high for a few, in both groups.

* Loan tenure shows some separation: defaulters peak more at longer tenures.

* Principal outstanding is right-skewed, but fairly similar between groups.

* Open and closed accounts show spikes at small integers.

* Total loan months shows slightly higher values for defaulters.

* Delinquent months are clearly higher for defaulters.

* Total DPD is also much larger for defaulters.

* Enquiry count is similar, centered around 4–6.

* Credit utilization shows strong separation: defaulters cluster near very high utilization.

* Overall, behavior variables (delinquency, DPD, utilization, tenure) distinguish defaulters more than demographic or loan-size variables.
